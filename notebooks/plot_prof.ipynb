```python
import h5py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
# import arya
import seaborn as sns
```

```python
from astropy.coordinates import SkyCoord
import astropy.units as u
import astropy.constants as c
```

```python
filename = "isolation/iso_scaled.hdf5"

f = h5py.File(filename, "r")
```

```python
Î¦ = f["PartType1/Potential"][()]
```

```python


pos = f[('PartType1/Coordinates')][()]
x = pos[:, 0]
y = pos[:, 1]
z = pos[:, 2]
r = np.sqrt(x**2 + y**2 + z**2)

vel = f[('PartType1/Velocities')][()]
vx = vel[:, 0]
vy = vel[:, 1]
vz = vel[:, 2]
v = np.sqrt(vx**2 + vy**2 + vz**2)
```

```python
plt.hist2d(x, y, norm=mpl.colors.LogNorm(10, 1e5), cmin=1, bins=50);
```

```python
plt.hist2d(vx, vz, norm=mpl.colors.LogNorm(10, 1e5), cmin=1, bins=50);
```

```python
plt.hist2d(r, v, norm=mpl.colors.LogNorm(10, 1e5), cmin=0, bins=100);
plt.xlabel("r")
plt.ylabel("v")
```

```python
plt.hist(r, 100, histtype="step")
plt.yscale("log")
plt.xlabel("R")
```
