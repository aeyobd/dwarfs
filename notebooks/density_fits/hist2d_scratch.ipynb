```julia
using Revise
```

```julia
import Arya
```

```julia
using GLMakie
```

```julia
x = -abs.(randn(10000))
y = -abs.(2randn(10000) )
```

```julia
h = Arya.histogram2d(x, y, (15, 10))
```

```julia

heatmap(h, axis=(; aspect=DataAspect())
)
```

```julia
h = Arya.histogram2d(x, y, (15, 10), normalization=:density)
```

```julia
sum(h.values .* Arya.bin_volumes(h.xbins, h.ybins))
```

```julia

heatmap(h, axis=(; aspect=DataAspect())
)
```

```julia
x, y = Arya.bin_centres(h)
```

```julia
scatter(vec(x), vec(y), color=vec(h.values))
```
