```julia
using OrderedCollections
```

```julia
function get_orbit_props(modelname)
    return read_fits(joinpath(modelname, "orbital_properties.fits"))
end
```

```julia
function compare_hists(orbit_props, key)
    fig = Figure()
    ax = Axis(fig[1,1], xlabel=string(key))

    for (label, props) in orbit_props
        if key ∈ names(props)
            stephist!(props[!, key], label=label)
        end
    end

    axislegend()
    fig
end
```
